Q2. What is the difference between the functions loc and iloc in pandas.DataFrame?

loc selects rows and columns with specific labels.   
iloc selects rows and columns at specific integer positions.     
Use loc for label-based indexing and iloc for integer-based indexing.

Q6. What do you understand about the windows function in pandas and list the types of windows
functions?

**What Are Window Functions?**     
Window functions (also known as rolling or expanding functions) operate on a sliding partition of data within a series or DataFrame.     
These functions perform an aggregation over a specified window of values.     
Think of it as a way to compute statistics over a moving subset of your data.

**Types of Window Functions in Pandas:**   
**Rolling Window:**   
A generic sliding window over the values.   
You specify the window size using the window parameter.   

```python
s = pd.Series(range(5))
s.rolling(window=2).sum()
```

**Weighted Window:**    
Uses non-rectangular, weighted windows (provided by the scipy.signal library).    
Useful for specialized calculations.

```python
# s.rolling(window='2D').sum()  # Sum over a 2-day window
```

**Expanding Window:**       
An accumulating window that grows with each observation.     
Computes cumulative statistics.

```python
df = pd.DataFrame({'A': ['a', 'b', 'a', 'b', 'a'], 'B': range(5)})
df.groupby('A').expanding().sum()
```

Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and
calculates the difference between them in days, hours, and minutes using Pandas time delta. The
program should prompt the user to enter the dates and display the result.

```python
import pandas as pd

def date_diff():
  """
  This function calculates the difference between two dates in days, hours, and minutes using Pandas time delta.
  """
  # Prompt the user to enter the dates
  date1 = input("Enter the first date (YYYY-MM-DD): ")
  date2 = input("Enter the second date (YYYY-MM-DD): ")

  # Convert the dates to Pandas datetime objects
  date1 = pd.to_datetime(date1)
  date2 = pd.to_datetime(date2)

  # Calculate the time delta between the dates
  time_delta = date2 - date1

  # Extract the days, hours, and minutes from the time delta
  days = time_delta.days
  hours = time_delta.seconds // 3600
  minutes = (time_delta.seconds % 3600) // 60

  # Display the result
  print(f"The difference between {date1} and {date2} is:")
  print(f"{days} days, {hours} hours, and {minutes} minutes")

# Call the function
date_diff()
```

Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified
column to a categorical data type. The program should prompt the user to enter the file path, column
name, and category order, and then display the sorted data.

```python
import pandas as pd

def convert_to_categorical():
  """
  This function reads a CSV file, converts a specified column to a categorical data type, and displays the sorted data.
  """
  # Prompt the user to enter the file path
  file_path = input("Enter the file path: ")

  # Prompt the user to enter the column name
  column_name = input("Enter the column name: ")

  # Prompt the user to enter the category order
  categories = input("Enter the category order (comma-separated): ").split(",")

  # Read the CSV file
  df = pd.read_csv(file_path)

  # Convert the specified column to a categorical data type
  df[column_name] = pd.Categorical(df[column_name], categories=categories, ordered=True)

  # Sort the data by the specified column
  df = df.sort_values(by=column_name)

  # Display the sorted data
  print(df)

# Call the function
convert_to_categorical()
```

Q11. You are given a CSV file containing student data that includes the student ID and their test score. Write
a Python program that reads the CSV file, calculates the mean, median, and mode of the test scores, and
displays the results in a table.


The program should do the following:
I Prompt the user to enter the file path of the CSV file containing the student data.   
I Read the CSV file into a Pandas DataFrame.    
I Calculate the mean, median, and mode of the test scores using Pandas tools.    
I Display the mean, median, and mode in a table.   

Assume the CSV file contains the following columns:

I Student ID: The ID of the student.   
I Test Score: The score of the student's test.   

Example usage of the program:    
Enter the file path of the CSV file containing the student data: student_data.csv

<pre>
+-----------+--------+
| Statistic | Value |
+-----------+--------+
| Mean | 79.6 |
| Median | 82 |
| Mode | 85, 90 |
+-----------+--------+

</pre>


Assume that the CSV file student_data.csv contains the following data:
Student ID,Test Score
<pre>
1,85

7,78

</pre>
The program should calculate the mean, median, and mode of the test scores and display the results
in a table.

```python
import pandas as pd

def main():
    # Prompt user for file path
    file_path = input("Enter the path to the CSV file containing student data: ")

    # Read the CSV file into a Pandas DataFrame
    try:
        df = pd.read_csv(file_path)
    except FileNotFoundError:
        print("File not found. Please provide a valid file path.")
        return

    # Calculate mean, median, and mode of test scores
    mean_score = df["Test Score"].mean()
    median_score = df["Test Score"].median()
    mode_scores = df["Test Score"].mode()

    # Display results in a table
    print("+-----------+--------+")
    print("| Statistic | Value  |")
    print("+-----------+--------+")
    print(f"| Mean      | {mean_score:.1f}   |")
    print(f"| Median    | {median_score:.0f}   |")
    print(f"| Mode      | {', '.join(map(str, mode_scores))} |")
    print("+-----------+--------+")

if __name__ == "__main__":
    main()
```
